```python
#Populating the dataframe
new_file= open("foods.txt", encoding = "ISO-8859-1")
lines = new_file.readlines()
#Creating an empty dictionary and appending it with every attribute
dictionary = {}
dictionary['product/productId'] = []
dictionary['review/userId'] = []
dictionary['review/profileName'] = []
dictionary['review/helpfulness'] = []
dictionary['review/score'] = []
dictionary['review/time'] = []
dictionary['review/summary'] = []
dictionary['review/text'] = []
```

```python
for line in lines:
  if len(line.split(":")) == 1:
    continue
  else :
    (key, value) = line.split(":",1)
    value = value.strip("\n")
    dictionary[key].append(value)
```

```python
#Transforming the dataframe
data = pd.DataFrame.from_dict(dictionary,orient='index').transpose()
```

**(i) Identify all the unique words that appear in the “review/text” field of the reviews. Denote the set of such words as L.**

```python
L = set()
unique = []
for index, rows in data.iterrows():
  words = set(re.split(r'[`!@#$%^&*()_+\-=\[\]{};\:"\\|,.<>\/?~1234567890 ]', data.loc[index,'review/text'].lower()))
  unique.append(words)
```

```python
merged = list(itertools.chain(*unique))
```

```python
L = set(merged)
```

**(ii) Remove from L all stopwords in “Long Stopword List” from http://www.ranks.nl/ stopwords. Denote the cleaned set as W.**

```python
stopwords_file = open('/content/stop_words.txt') #Stored and uploaded the set of stopwords from a file
stopwords = stopwords_file.readlines()
stopwords = [stopword.strip("\\\n") for stopword in stopwords]

W = [w for w in L if w not in stopwords] #Cleaned set W after removing stopwords
```

```python
print(W) #set of cleaned words
```

**(iii) Count the number of times each word in W appears among all reviews (“review/text” field) and identify the top 500 words.**

```python
count = collections.Counter(merged)
countin_w = {}
for word in W:
  countin_w[word] = count[word]
```

```python
frequency = {k: v for k, v in sorted(countin_w.items(), key=lambda item: item[1],reverse = True)}
top_500 = {k: frequency[k] for k in list(frequency)[1:501]} #Sorted in descending order to get the top 500 words
```

```python
list_500 = list(top_500.keys())
```
